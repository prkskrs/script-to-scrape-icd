```python
def chunkIt(seq, num):
    avg = len(seq) / float(num)
    out = []
    last = 0.0

    while last < len(seq):
        out.append(seq[int(last):int(last + avg)])
        last += avg

    return out
```

```python
a=chunkIt(uhidd, 2)
a
b=chunkIt(treating_doctorr, 2)
b
c=chunkIt(asst_surgeonn, 2)
c
d=chunkIt(condition_during_dischargee, 2)
d
```
